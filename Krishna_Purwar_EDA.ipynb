```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the datasets from CSV files
df_customers = pd.read_csv("https://raw.githubusercontent.com/i183x/zeo/refs/heads/main/Customers.csv")
df_products = pd.read_csv("https://raw.githubusercontent.com/i183x/zeo/refs/heads/main/Products.csv")
df_transactions = pd.read_csv("https://raw.githubusercontent.com/i183x/zeo/refs/heads/main/Transactions.csv")
```

```python
# Display the first few rows of each dataset
print("Customers DataFrame:")
print(df_customers.head())
print("\nProducts DataFrame:")
print(df_products.head())
print("\nTransactions DataFrame:")
print(df_transactions.head())
```

```python
# Display column names for each DataFrame
print("Column names of Customers:")
print(df_customers.columns)
print("Column names of Products:")
print(df_products.columns)
print("Column names of Transactions:")
print(df_transactions.columns)
```

```python
# Check for missing values in each dataset
print("\nCustomers Missing Values:")
print(df_customers.isnull().sum())
print("\nProducts Missing Values:")
print(df_products.isnull().sum())
print("\nTransactions Missing Values:")
print(df_transactions.isnull().sum())
```

```python
# Check the data types and basic info of each DataFrame
print("\nCustomers Info:")
print(df_customers.info())

print("\nProducts Info:")
print(df_products.info())

print("\nTransactions Info:")
print(df_transactions.info())
```

```python
# Convert 'SignupDate' and 'TransactionDate' to datetime format
df_customers['SignupDate'] = pd.to_datetime(df_customers['SignupDate'])
df_transactions['TransactionDate'] = pd.to_datetime(df_transactions['TransactionDate'])

# Summary statistics for numeric columns in Customers, Products, and Transactions datasets
print(df_customers.describe())
```

```python
# Customer Insights
# 1.1. Customer Distribution by Region

# Plot number of customers by region using countplot
plt.figure(figsize=(10, 6))
sns.countplot(data=df_customers, x='Region', palette='Set2')
plt.title('Customer Count by Region')
plt.show()

# Pie chart for customer distribution by region
plt.figure(figsize=(10, 6))
plt.pie(df_customers['Region'].value_counts(), labels=df_customers['Region'].unique(), autopct='%1.1f%%')
plt.title('Customer Count by Region')
plt.show()

# Summary statistics for numeric columns in the Products dataset
print(df_products.describe())
```

```python
# 1.2. Product Distribution by Category

# Plot product count by category using countplot
plt.figure(figsize=(10, 6))
sns.countplot(data=df_products, x='Category', palette='Set3')
plt.title('Product Count by Category')
plt.xticks(rotation=45)
plt.show()

# Pie chart for product distribution by category
plt.figure(figsize=(10, 6))
plt.pie(df_products['Category'].value_counts(), labels=df_products['Category'].unique(), autopct='%1.1f%%', textprops={'fontsize': 10})
plt.title('Product Count by Category')
plt.show()

# Summary statistics for numeric columns in Transactions dataset
print(df_transactions.describe())

```

```python
# 1.3. Distribution of Transaction Values

# Plot the distribution of total transaction values
plt.figure(figsize=(10, 6))
sns.histplot(df_transactions['TotalValue'], kde=True, bins=50, color='green')
plt.title('Distribution of Total Transaction Value')
plt.show()

# Plot the distribution of quantities purchased
plt.figure(figsize=(10, 6))
sns.histplot(df_transactions['Quantity'], kde=True, bins=50, color='orange')
plt.title('Distribution of Quantity Purchased')
plt.show()

# Drop the 'Price' column from transactions as it is a duplicate
df_transactions = df_transactions.drop('Price', axis=1)
```

```python
# Merge the datasets (Customers, Products, and Transactions)
df = df_transactions.merge(df_customers, on='CustomerID', how='left')
df = df.merge(df_products, on='ProductID', how='left')

# Preview the merged DataFrame
print("Combined dataframe:")
print(df.head())
```

```python
# Sales Insights
# 2.1. Total Sales by Region

# Group the sales data by region and calculate the total sales
sales_by_region = df.groupby('Region')['TotalValue'].sum().reset_index()

# Plot total sales by region using barplot
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='TotalValue', data=sales_by_region, palette='coolwarm')
plt.title('Total Sales by Region')
plt.show()

# Pie chart for total sales by region
plt.figure(figsize=(10, 6))
plt.pie(sales_by_region['TotalValue'], labels=sales_by_region['Region'], autopct='%1.1f%%')
plt.title('Total Sales by Region')
plt.show()

```

```python
# 2.2. Total Sales by Product Category

# Group the sales data by product category and calculate total sales
sales_by_category = df.groupby('Category')['TotalValue'].sum().reset_index()

# Plot total sales by category using barplot
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='TotalValue', data=sales_by_category, palette='Blues')
plt.title('Total Sales by Product Category')
plt.xticks(rotation=45)
plt.show()

# Pie chart for total sales by category
plt.figure(figsize=(10, 6))
plt.pie(sales_by_category['TotalValue'], labels=sales_by_category['Category'], autopct='%1.1f%%', textprops={'fontsize': 10})
plt.title('Total Sales by Product Category')
plt.show()

```

```python
# 2.3. Customer Signups Over Time

# Create a new column for Year-Month from the SignupDate
df_customers['YearMonth'] = df_customers['SignupDate'].dt.to_period('M').astype(str)

# Group the customers by signup month and count the number of customers
signup_by_month = df_customers.groupby('YearMonth')['CustomerID'].count().reset_index()

# Plot the customer signups over time using a lineplot
plt.figure(figsize=(10, 6))
sns.lineplot(x='YearMonth', y='CustomerID', data=signup_by_month, marker='o', color='purple')
plt.title('Customer Signups Over Time')
plt.xticks(rotation=45)
plt.show()

# Sort the signup data in ascending order
signup_by_month_sorted = signup_by_month.sort_values(by='CustomerID', ascending=True)

# Display the sorted data
print(signup_by_month_sorted)

# Find the month with the highest signups
max_signup_month = signup_by_month_sorted.iloc[-1]

# Print the month with the highest signups
print("Month with the highest signup:")
print(max_signup_month)

# Highlight the highest signup month in the plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='YearMonth', y='CustomerID', data=signup_by_month, marker='o', color='purple')

# Highlight the max signup month
plt.scatter(max_signup_month['YearMonth'], max_signup_month['CustomerID'], color='red', marker='o', s=100)

# Annotate the highest signup month
plt.annotate(f"Max Signup: {max_signup_month['CustomerID']}",
             (max_signup_month['YearMonth'], max_signup_month['CustomerID']),
             textcoords="offset points",
             xytext=(0, 10), ha='center')

# Set title and rotate xticks
plt.title('Customer Signups Over Time')
plt.xticks(rotation=45)
plt.show()

```

```python
# 2.4. Monthly Signups

# Extract the month name from the 'YearMonth' column
signup_by_month['Month'] = pd.to_datetime(signup_by_month['YearMonth']).dt.month_name()

# Group by month and calculate the total signups for each month
monthly_totals = signup_by_month.groupby('Month')['CustomerID'].sum().reindex([
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
])

# Find the best month (highest signups)
best_month = monthly_totals.idxmax()
best_month_total = monthly_totals.max()

# Plot the total signups by month using a barplot
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_totals.index, y=monthly_totals.values, palette='viridis')

# Highlight the best month
plt.bar(best_month, best_month_total, color='red')

# Annotate the best month
plt.text(
    best_month, best_month_total + 50, f"{best_month_total}",
    ha='center', color='black', fontsize=12
)

# Set plot title and labels
plt.title('Total Signups by Month (Combined Over Two Years)')
plt.xlabel('Month')
plt.ylabel('Total Signups')
plt.xticks(rotation=45)
plt.show()

# Print the best month details
print(f"The best month is {best_month} with {best_month_total} total signups.")

# 2.5. Pie Chart of Signups by Month
plt.figure(figsize=(8, 8))
colors = sns.color_palette('viridis', len(monthly_totals))

# Plot the pie chart for signups by month
plt.pie(
    monthly_totals.values, labels=monthly_totals.index, autopct='%1.1f%%', startangle=90, colors=colors,
    wedgeprops={'edgecolor': 'black'}
)

# Title for the pie chart
plt.gca().set_title(f"Best Month: {best_month} ({best_month_total} Signups)", fontsize=14, weight='bold')

# Show the pie chart
plt.tight_layout()
plt.show()

# Print the best month details again
print(f"The best month is {best_month} with {best_month_total} total signups.")

```

```python
# Transaction Insights
# 3.1. Transaction Count Over Time

# Create a new column for Year-Month from the TransactionDate
df_transactions['YearMonth'] = df_transactions['TransactionDate'].dt.to_period('M').astype(str)

# Group the transactions by month and count the number of transactions
transactions_by_month = df_transactions.groupby('YearMonth')['TransactionID'].count().reset_index()

# Plot the transaction count over time using a lineplot
plt.figure(figsize=(10, 6))
sns.lineplot(x='YearMonth', y='TransactionID', data=transactions_by_month, marker='o', color='blue')
plt.title('Transaction Count Over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlations
# 4.1. Correlation Matrix

# Calculate the correlation between numeric columns: Quantity, Price, and TotalValue
corr_matrix = df[['Quantity', 'Price', 'TotalValue']].corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Customer Behavior
# 5.1. Repeat vs One-time Customers

# Group by CustomerID to count the number of transactions per customer
repeat_customers = df.groupby('CustomerID')['TransactionID'].count()

# Count how many customers made more than one purchase
repeat_customers_count = repeat_customers[repeat_customers > 1].count()

# Calculate the percentage of one-time customers
percentage_one_time_customers = (repeat_customers[repeat_customers == 1].count() / len(repeat_customers)) * 100

# Print repeat and one-time customer statistics
print(f"Repeat customers: {repeat_customers_count}")
print(f"One-time customers: {percentage_one_time_customers:.2f}%")
```

```python
# 5.2. Region-Category Sales Interaction

# Group the sales data by Region and Category to calculate total sales
region_category = df.groupby(['Region', 'Category'])['TotalValue'].sum().unstack()

# Plot the sales interaction between Region and Category using a heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(region_category, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Sales by Region and Category")
plt.show()
```

```python
# 5.3. Time-to-First Purchase

# Calculate the time (in days) from SignupDate to the first purchase for each customer
df['FirstPurchase'] = df.groupby('CustomerID')['TransactionDate'].transform('min')
df['DaysToFirstPurchase'] = (df['FirstPurchase'] - df['SignupDate']).dt.days

# Plot the distribution of days to first purchase using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['DaysToFirstPurchase'].dropna(), kde=True, bins=30, color='teal')
plt.title('Days Between Signup and First Purchase')
plt.xlabel('Days')
plt.ylabel('Frequency')
plt.show()
```

```python
# Sales Performance
# 6.1. Category Growth Over Time

# Group sales data by month and product category, then calculate the total sales for each category each month
category_trend = df.groupby([df['TransactionDate'].dt.to_period('M'), 'Category'])['TotalValue'].sum().unstack()

# Plot the sales growth by category over time
plt.figure(figsize=(12, 6))
category_trend.plot(kind='line', marker='o', figsize=(12, 6), colormap='viridis')
plt.title('Monthly Sales Trends by Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# 6.2. High-Value Transactions

# Calculate the 90th percentile for the TotalValue column
high_value_threshold = df['TotalValue'].quantile(0.9)

# Calculate the percentage of transactions above the 90th percentile
high_value_percentage = (df[df['TotalValue'] > high_value_threshold].shape[0] / df.shape[0]) * 100

# Print the percentage of high-value transactions
print(f"Percentage of high-value transactions: {high_value_percentage:.2f}%")

```

```python
# Operational Efficiency
# 7.1. Order Quantity Trends

# Plot the average order quantity over time
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['TransactionDate'].dt.to_period('M').astype(str), y=df['Quantity'])
plt.title('Average Order Quantity Over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
# 7.2. Region-Wise Efficiency

# Calculate the average total value per transaction for each region
avg_value_by_region = df.groupby('Region')['TotalValue'].mean().sort_values()

# Plot the average transaction value by region
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_value_by_region.index, y=avg_value_by_region.values, palette='mako')
plt.title('Average Total Value per Transaction by Region')
plt.ylabel('Average Transaction Value')
plt.xticks(rotation=45)
plt.show()
```

```python
# Customer Lifecycle Value
# 8.1. Lifetime Value (LTV)

# Calculate the total revenue generated by each customer
customer_ltv = df.groupby('CustomerID')['TotalValue'].sum()

# Plot the distribution of customer lifetime value
plt.figure(figsize=(10, 6))
sns.histplot(customer_ltv, kde=True, bins=30, color='gold')
plt.title('Customer Lifetime Value Distribution')
plt.xlabel('Total Revenue Generated')
plt.ylabel('Frequency')
plt.show()

```
